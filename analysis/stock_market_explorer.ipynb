```python
import pandas as pd
import yfinance as yf
```

```python
TICKERS = [
    # Mega-cap / Big Tech
    "AAPL", "MSFT", "AMZN", "GOOG", "META", "NVDA", "TSLA",

    # Semis / Hardware
    "AMD", "INTC", "QCOM", "AVGO", "TXN", "MU", "AMAT", "LRCX",

    # Cloud / Software
    "ORCL", "CRM", "ADBE", "NOW", "INTU",

    # Streaming / Media
    "NFLX", "DIS",

    # Payments / Fintech
    "V", "MA", "PYPL",

    # Banks / Financials
    "JPM", "BAC", "WFC", "GS", "MS", "C",

    # Consumer / Retail
    "COST", "WMT", "TGT", "NKE", "SBUX", "MCD",

    # Health
    "UNH", "JNJ", "PFE", "ABBV", "MRK",

    # Energy
    "XOM", "CVX",

    # Industrials / Other
    "BA", "CAT", "GE",

    # “Market proxies” (ETFs)
    "SPY", "QQQ", "DIA", "IWM",
]
START = "2019-01-01"

raw = yf.download(
    tickers=TICKERS,
    start=START,
    auto_adjust=True,
    progress=False
)

raw.head()
```

```python
close = raw["Close"].copy()

df = close.reset_index().melt(
    id_vars=["Date"],
    var_name="Ticker",
    value_name="Close"
).sort_values(["Ticker", "Date"])

df["DailyReturn"] = df.groupby("Ticker")["Close"].pct_change()
df["Date"] = pd.to_datetime(df["Date"])

df.head()
```

```python
df.to_csv("../public/data/stocks_clean.csv", index=False)
```
